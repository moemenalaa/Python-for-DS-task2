### Problem Descripition 

In 2012, URL shortening service Bitly partnered with the US government website USA.gov to provide a feed of anonymous data gathered from users who shorten links ending with .gov or .mil.

The text file comes in JSON format and here are some keys and their description. They are only the most important ones for this task.

|key| description |
|---|-----------|
| a|Denotes information about the web browser and operating system|
| tz | time zone |
| r | URL the user come from |
| u | URL where the user headed to |
| t | Timestamp when the user start using the website in UNIX format |
| hc | Timestamp when user exit the website in UNIX format |
| cy | City from which the request intiated |
| ll | Longitude and Latitude |

In the cell, I tried to provide some helper code for better understanding and clearer vision

```python
# I will try to retrieve one instance of the file in a list of dictionaries
import json
import re
records = [json.loads(line) for line in open('usa.gov_click_data.json')]
# Print the first occurance
print(type(records))
records[0]
```

```python
import json 
import pandas as pd 
from pandas.io.json import json_normalize 

df = json_normalize(records)
```

```python
df.head()
```

```python
df = df.drop(columns=['_heartbeat_','al','c','g','gr','h','hh','kw','l','nk'])
```

```python
df.dropna(inplace=True)
```

- web_browser
        The web browser that has requested the service
- operating_sys
        operating system that intiated this request

```python
def find_os(txt):
    return re.findall(r'\(.*?\)', txt)

    
```

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = "," 
    
    # return string   
    return (str1.join(s)) 
```

```python
df['operating_sys'] = df['a'].apply(find_os)
```

```python
result = []
for a,b in zip(df['a'], df['operating_sys']):
    for i in b:
        a = a.replace(i,"")
    result.append(a)   
```

```python
df['a'] = pd.DataFrame(result)
```

```python
df.rename(columns={'a':'web_browser'},inplace=True)
```

```python
df.head()
```

```python
def replace_p(txt):
    txt = txt.replace('(','')
    txt = txt.replace(')','')
    return txt
```

```python
def xstr(s):
    if s is None:
        return ''
    return str(s)
```

```python
df['operating_sys'] = df['operating_sys'].apply(listToString)
```

```python
df['operating_sys'] = df['operating_sys'].apply(replace_p)
```

```python
df.head()
```

- city

        The city from which the the request was sent

```python
df = df.rename(columns={'cy':'city'})
```

from_url

  The main URL the user came from


to_url

 The same applied like `to_url`

```python
def split(strng):
    strng = strng.split('/')
    return '/'.join(strng[:3])
```

```python
df['r'] = df['r'].apply(split)
```

```python
df = df.rename(columns={'r':'from_url'})
```

```python
df['u'] = df['u'].apply(split)
```

```python
df = df.rename(columns={'u':'to_url'})
```

longitude

  The longitude where the request was sent
  
latitude

  The latitude where the request was sent

```python
df[['longitude','latitude']] = pd.DataFrame(df.ll.tolist(), index= df.index)
```

```python
df.drop(columns=['ll'],inplace=True)
```

- time_zone
        
        The time zone that the city follow

```python
df = df.rename(columns={'tz':'time_zone'})
```

- time_in

        Time when the request started

```python
df = df.rename(columns={'t':'time_in'})
```

- time_out
        
        Time when the request is ended

```python
df = df.rename(columns={'hc':'time_out'})
```

```python
df.head()
```

```python
df = df[['web_browser','operating_sys','from_url','to_url','city','longitude','latitude','time_zone','time_in','time_out']]
```

```python
df.head()
```

```python
df.to_csv('out.csv', index=False)
```
